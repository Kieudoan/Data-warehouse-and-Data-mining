# Tiền xử lý dữ liệu cho phân cụm 

```python
import pandas as pd
dataframe = pd.read_csv('Nhom 8_Data4KNN_ID3_Kmean_Hierarchical.csv')
```

```python
dataframe.info()
```

```python
special_characters = ['?', '!', '#',"'"]

# Hàm để loại bỏ các kí tự đặc biệt
def remove_special_chars(text):
    for char in special_characters:
        text = text.replace(char, '')
    return text

# Loại bỏ các kí tự đặc biệt bằng hàm remove_special_chars
dataframe['Surname'] = dataframe['Surname'].apply(remove_special_chars)

# Hiển thị dữ liệu sau khi loại bỏ
print(dataframe.head())
```

```python
# Xóa dữ liệu trùng
dataframe=dataframe.drop_duplicates()
print(dataframe)
```

```python
# Loại bỏ oulier
columns=['CreditScore','Age','Tenure','Balance','EstimatedSalary'] 
for column in columns: 
    q1 = dataframe[column].quantile(0.25)
    q3 = dataframe[column].quantile(0.75)
    iqr = q3 - q1 # tính IQR
    dataframe = dataframe.loc[(dataframe[column] > q1 - 1.5 * iqr) & (dataframe[column] < q3 + 1.5 * iqr)]  
dataframe.reset_index(drop=True,inplace=True)
print(dataframe)
```

```python
# Giảm số bản ghi còn 300 dòng
cluster_df = dataframe[:301]
print(cluster_df.head())

# Xuất file csv đã tiền xử lý phân cụm (k-means, hierarchical) để sử dụng trong orange
cluster_df.to_csv('cluster_data.csv')
```

# Tiền xử lý dữ liệu cho phân lớp

```python
import pandas as pd
# Đọc file CSV
df = pd.read_csv('Nhom 8_Data4KNN_ID3_Kmean_Hierarchical.csv')
df
```

```python
# xóa các cột không cần thiết
df=df.drop(columns=['CustomerId', 'Surname'],axis=1)
df
```

```python
# Đổi 1,0 thành Yes, No cho cột exited
# Tạo dictionary để map giá trị số sang chữ
lookup_table = {
    0: "No",
    1: "Yes"
}
df["Exited"] = df["Exited"].map(lookup_table)
# Hiển thị kết quả
print(df.to_string())
```

```python
# Tạo các khoảng cho cột Age
# Khởi tạo hàm ánh xạ
def age_group(age):
    if age <= 30:
        return "<30"
    elif age > 30 and age <= 60 :
        return "31-60"
    else:
        return ">60"

# Ánh xạ giá trị tuổi sang nhãn nhóm tuổi
age_groups = df["Age"].map(age_group)
# Thêm cột mới vào DataFrame
df["Age_Group"] = age_groups
df
```

```python
# Chia cột "Balance" thành 3 khoảng với nhãn
balance_groups = pd.cut(df["Balance"], bins=3, labels=["Low", "Medium", "High"])
# Thêm cột mới vào DataFrame
df["Balance_Group"] = balance_groups
# Hiển thị dữ liệu với cột "Balance_Group"
print(df)
```

```python
# Ghi dữ liệu vào file csv
df.to_csv('bank.csv')
```
